## 2) Load training dataset

*  Read from Keras library.



```
# Load the Pandas libraries with alias 'pd' 
import pandas as pd 
import numpy as np

# Read data 
from keras.datasets import mnist
(trainData, trainLabels), (testData, testLabels) = mnist.load_data()

# Reshape training to 1D
train_sz = trainData.shape
trainSet = np.reshape(trainData, (train_sz[0], train_sz[1]*train_sz[2])).astype('float32')

# Reshape testing to 1D
test_sz = testData.shape
testSet = np.reshape(testData, (test_sz[0], test_sz[1]*test_sz[2])).astype('float32')

# Normalize
from sklearn.preprocessing import normalize
trainSet = normalize(trainSet)
testSet = normalize(testSet)

# [Optional:] Check the dimentionality and type of your training and testing data
print('Training Data:', trainSet.shape)
print('Testing Data:', testSet.shape)
tdatadf=pd.DataFrame(trainSet)
tlabeldf=pd.DataFrame(trainLabels)
tdf=pd.concat([tdatadf,tlabeldf])
```

##4) Shuffle and Split TRAINING data as train (also called development) (80%) and validation (20%) 

```
from sklearn.utils import shuffle

# Shuffle the training data

trainSet,trainLabels=shuffle(trainSet,trainLabels)





# Split 80-20
from sklearn.model_selection import train_test_split

(train_x, val_x, train_y, val_y) = train_test_split(trainSet, trainLabels, test_size=0.2)




```

##5) Train decision tree and k-NN  classifiers on development data and do model selection using the validation data

* Train a decision tree classifier with different values of "min_samples_split" which is the minimum number of samples required to split an internal node:  min_samples_split = [default = 2, 3, 5]. 

* Train a k-NN classifier (k=1 and k=5 and rest of the parameters set to default). 

```

# Train decision tree classifiers
from sklearn.tree import DecisionTreeClassifier
samplesplits=[2,3,5]
for l in samplesplits:
  dt = DecisionTreeClassifier(min_samples_split=l)
  dt = dt.fit(train_x, train_y)

  test_pred = dt.predict(val_x)

  from sklearn.metrics import accuracy_score
  TestAcc = accuracy_score(val_y, test_pred)*100
  print(TestAcc)





# Train k-NN classifiers
from sklearn.neighbors import KNeighborsClassifier
kVals = [1,5]

accuracies = []

# loop over values of k for the k-Nearest Neighbor classifier

for k in kVals:
  
  model = KNeighborsClassifier(n_neighbors=k)
  model.fit(train_x, train_y)
  
  # Evaluate the model on validation set 
  score = model.score(val_x, val_y)
  print("For k = %d, validation accuracy = %.5f%%" % (k, score * 100))
  
  # Update the accuracies list
  accuracies.append(score)




```
